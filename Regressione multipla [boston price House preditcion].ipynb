costruire un modello di regressione lineare multipla

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

eseguiamo una regressione lineare multipla su queste features

```python
# creiamo due np array

X= df[["RM","LSTAT"]].values
# X= proprietà
Y= df["MEDV"].values
# Y= target
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=0)
```

```python
# costruiamo il modello:

ll = LinearRegression() 
# regressione

ll.fit(X_train, Y_train) 
# fit

Y_pred = ll.predict(X_test) 
#predizione
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(Y_test, Y_pred)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Y_test, Y_pred)
```

Eseguiamo una regressione lineare multipla su tutte le features

```python
from sklearn.preprocessing import StandardScaler

# Questa volta standardizziamo il dataset visto che abbiamo
# molte proprietà in scale differenti
```

```python
X = df.drop("MEDV", axis=1).values
Y = df["MEDV"].values

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=0)
```

```python
ss = StandardScaler()
X_train_std = ss.fit_transform(X_train)
X_test_std = ss.transform(X_test)
```

```python
ll = LinearRegression()
ll.fit(X_train_std, Y_train)
Y_pred = ll.predict(X_test_std)
```

```python
print("r2_score: " +str(r2_score(Y_test, Y_pred)))
print("MSE: " +str(mean_squared_error(Y_test, Y_pred)))
```
