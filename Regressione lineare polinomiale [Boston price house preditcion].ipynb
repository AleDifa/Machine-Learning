```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=0)
```

```python
# per eseguire una regressione polinomiale con Sklern
#  dobiamo creare delle proprietà aggiuntivi con la classe:
#   PolynomialFeatures
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
#istanza di PolyFeatures e dentro le ( ) il grado 
```

```python
PolyFeatures=PolynomialFeatures(degree=2)
```

```python
#creare nuovi array con le Features
```

```python
X_train_poly = PolyFeatures.fit_transform(X_train)
X_test_poly = PolyFeatures.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
ll = LinearRegression()
```

```python
ll.fit(X_train_poly, Y_train)
```

```python
Y_pred = ll.predict(X_test_poly)
```

vediamo quanto è valido questo modello

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(Y_test, Y_pred)
print(mse)
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(Y_test, Y_pred)
print(r2)
```

```python
# Quanti gradi servono per un risulato migliore?
# creiamo un ciclo che provi fino al 10° gradp
```

```python
scores = []

for i in range(1,11):
    # (1,11) specifichiamo fino al 10° grado
    polyfeats = PolynomialFeatures(degree=i)
    X_train_poly = polyfeats.fit_transform(X_train)
    X_test_poly = polyfeats.transform(X_test)
    
    ll = LinearRegression()
    ll.fit(X_train_poly, Y_train)
    Y_pred = ll.predict(X_test_poly)
    
    mse = mean_squared_error(Y_test, Y_pred)
    r2 = r2_score(Y_test, Y_pred)
    scores.append(r2)
    
    print("DEGREE "+str(i)+": MSE="+str(mse)+" R2="+str(r2))
```

```python
# fino al DEGREE 4 migliora (r2)
# R2=0.6415805408226356
```

```python
# il peggioramento è dato dall'Overfitting
```

proviamo ad utilizzare tutte le proprietà per eseguire una regressione Polinomiale

```python
# riscriviamo tutto
```

```python
X=df.drop("MEDV", axis=1).values
```

```python
Y=df["MEDV"].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=0)
```

```python
scores = []

for i in range(1,5):
    # fino al 4° grado
    polyfeats = PolynomialFeatures(degree=i)
    X_train_poly = polyfeats.fit_transform(X_train)
    X_test_poly = polyfeats.transform(X_test)
    
    ll = LinearRegression()
    ll.fit(X_train_poly, Y_train)
    Y_pred = ll.predict(X_test_poly)
    
    mse = mean_squared_error(Y_test, Y_pred)
    r2 = r2_score(Y_test, Y_pred)
    scores.append(r2)
    print("DEGREE "+str(i)+": MSE="+str(mse)+" R2="+str(r2))
```
